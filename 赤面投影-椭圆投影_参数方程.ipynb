```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def pf(theta):
    t = np.linspace(0,2*np.pi,1000)
    r0 = np.sin(np.deg2rad(theta))
    y0 = np.cos(np.deg2rad(theta))
    
    x = 2*r0*np.cos(t)/(2*r0*np.sin(t)+1)
    y = 2*y0/(2*r0*np.sin(t)+1)

    plt.title('Parameter Equation Plot')
    plt.xlabel('x(t)')
    plt.ylabel('y(t)')
    return plt.plot(x,y)
```
